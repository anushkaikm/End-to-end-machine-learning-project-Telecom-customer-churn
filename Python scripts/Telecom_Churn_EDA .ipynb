We will now divide the consumers into 3 categories:



1.   **No plan** : Customers who have not opted for any service plan

1.   **Premium** : Customers who have opted for both the service plans








```python
# initialize list of lists
data = [['No plan', df.value_counts(["Voice mail plan", "International plan"])[0] ], ['Prospective', df.value_counts(["Voice mail plan", "International plan"])[1] + df.value_counts(["Voice mail plan", "International plan"])[2] ], ['Premium', df.value_counts(["Voice mail plan", "International plan"])[3] ]]
  
# Create the pandas DataFrame
cust_df = pd.DataFrame(data, columns=['Customer category', 'Number of customers'])
  
# print dataframe.
cust_df
```

```python
# initialize list of lists
subdata = [['Voice mail only', df.value_counts(["Voice mail plan", "International plan"])[1]], ['International only', df.value_counts(["Voice mail plan", "International plan"])[2]]]
  
# Create the pandas DataFrame
subcust_df = pd.DataFrame(subdata, columns=['Customer sub-category', 'Number of customers'])
  
# print dataframe.
subcust_df
```

```python
churn_value = df.groupby(['Voice mail plan','International plan','Churn']).size().reset_index(name='Churn_value',level=0)
churn_value
```

```python
churn_value2 = churn_value.groupby(["Voice mail plan", "International plan", "Churn"]).agg({"Churn_value" : "sum"})
churn_value2
```

```python
churn_value3 = df.groupby(['Voice mail plan','International plan','Churn']).size()
```

```python
churn_value3
```

```python
churn_value4 = round((churn_value3 / churn_value3.groupby(["International plan", "Voice mail plan"]).sum())*100)
```

```python
churn_percent = churn_value4.reset_index(name='Churn_%')
```

```python
churn_percent
```

```python
# initialize list of lists
data = [['No plan', 'No', 'No'], ['Voice mail only', 'Yes', 'No'], ['International only', 'No', 'Yes'], ['Premium', 'Yes', 'Yes']]
  
# Create the pandas DataFrame
cust_category_df = pd.DataFrame(data, columns=['Customer category', 'Voice mail plan', 'International plan'])
  
# print dataframe.
cust_category_df
```

```python
cust_category_df2 = pd.merge(churn_percent, cust_category_df, on=['Voice mail plan', 'International plan'], how = 'inner')
cust_category_df2
```

```python
# shift column 'Customer category' to first position
first_column = cust_category_df2.pop('Customer category')
  
# insert column using insert(position,column_name,
# first_column) function
cust_category_df2.insert(0, 'Customer category', first_column)
  
print()
print("After Shifting column to first position")
display(cust_category_df2)
```

So far, we have figured out the churn percentage of each category

```python
df1 = cust_category_df2.set_index(['Churn'])
churn_services_df = df1.loc[True]
churn_services_df
```

Now, we have exclusive information of percentage of customers that have churned from each customer category.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
fig = plt.figure(figsize=(12,5))
ax = fig.add_axes([0,0,1,1])

X = churn_services_df['Customer category']

Y = np.array(churn_services_df['Churn_%'])
ax.bar(X,Y)

for bar in ax.patches:
  # Using Matplotlib's annotate function and passing the coordinates where the annotation shall be done
  # x-coordinate: bar.get_x() + bar.get_width() / 2
  # y-coordinate: bar.get_height()
  # free space to be left to make graph pleasing: (0, 8)
  # ha and va stand for the horizontal and vertical alignment
    ax.annotate(format(bar.get_height(), '.2f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha = 'center', va = 'center', 
                   size=15,xytext = (0, 8), 
                   textcoords = 'offset points')

plt.xlabel("Customer category")
plt.ylabel("Churn %")
plt.title("Customer churn analysis in respective services category")
plt.show()
```

From this graph, we derive the following conclusions according by evaluating services attributes:


1.   Out of 2180 'No plan' category customers, 14% have churned.

## Histogram

```python
import plotly.express as px
```

```python
# Defining the histogram plotting function
def hist(feature):
    group_df = df.groupby([feature, 'Churn']).size().reset_index()
    group_df = group_df.rename(columns={0: 'Count'})
    fig = px.histogram(group_df, x=feature, y='Count', color='Churn', marginal='box', title=f'Churn rate frequency to {feature} distribution',
                       color_discrete_sequence=['yellow',"blue"])
    
    fig.show()

    plt.axvline(df[feature].mean(),color='orange',label='Mean')
    plt.axvline(df[feature].median(),color='black',label='Median')
    plt.show()
```

```python
hist('Account length')
print('Skewness of Account length:',df['Account length'].skew())
```

The 'Account length' histogram is barely skewed and shows that most customers have been with the telecom company for just the first quarter (120 days), especially from the last week of March to the end of April. The highest rate of churn is also in the first quarter (120 days), especially from the last week of March to the end of April. 

```python
hist('Total day charge')
print('Skewness of Total day charge:',df['Total day charge'].skew())
```

The Total day charge histogram shows that clients with higher day charges have a higher churn rate. Also, the histogram is negatively skewed (left-skewed).

```python
hist('Total eve charge')
print('Skewness of Total eve charge:',df['Total eve charge'].skew())
```

The Total eve charge histogram shows that clients with higher evening charges (13.5 - 18.49) have a higher churn rate. Also, the histogram is negatively skewed (left-skewed).

```python
hist('Total night charge')
print('Skewness of Total night charge:',df['Total night charge'].skew())
```

The Total night charge histogram shows that clients with higher night charges (7.5-10.49) have a higher churn rate. Also, the histogram is positively skewed (right-skewed).

```python
hist('Total intl charge')
print('Skewness of Total intl charge:',df['Total intl charge'].skew())
```

The Total intl charge histogram shows that clients with higher international charges (2.25-3.24) have a higher churn rate. Also, the histogram is negatively skewed (left-skewed).

# Multivariate Visualization

Multivariate plots allow us to see relationships between two and more different variables, all in one figure. Just as in the case of univariate plots, the specific type of visualization will depend on the types of the variables being analyzed.

**Median Value Comparison**


Median Value Comparison for Customers who churned vs not churned

```python
cx = df.groupby('Churn').median().transpose().plot(kind='barh', figsize = (15,15));
for p in cx.patches:
    width = p.get_width()
    plt.text(5+p.get_width(), p.get_y()+0.55*p.get_height(),
             '{:1.2f}'.format(width),
             ha='center', va='center')
plt.xlabel('Median Values')
plt.title('Median Values Comparion for Churn');

plt.figure(figsize = (16,12));
median_stats = df.groupby('Churn').median().transpose()
median_stats['%Difference Median'] = 100*((median_stats[True]-median_stats[False])/(median_stats[False]))
median_stats.sort_values(by='%Difference Median', inplace=True)
ax = median_stats['%Difference Median'].plot(kind='barh')
for p in ax.patches:
    width = p.get_width()
    plt.text(5+p.get_width(), p.get_y()+0.55*p.get_height(),
             '{:1.2f}'.format(width),
             ha='center', va='center')
plt.xlabel('Percentage Difference')
plt.title('Median Values %Difference')
```

**Analysis Summary:**

Looking at the median values of each feature and the percentage difference in median values for each category we find that Customers who churned:

1.   Made twice as much Customer service calls than those who didn't churn
2.   Had 22% more Total Day Charge

**Observation:**

Median values % Differences are same for the following pairs:

1.   Total day minutes and Total day charge
2.   Total eve minutes and Total eve charge
3.   Total night minutes and Total night charge
4.   Total int minutes and Total int charge




This suggests that minutes and charge columns are possibly strongly correlated. We will explore this more next when we look at correlations
